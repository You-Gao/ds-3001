1. What is the difference between regression and classification?

The difference of regression and classiifcation is that in regression you are predicting a value while in classification you are trying to predict the class of an observation. That is regression deals with continuous values while classification deals with binary values.

2. What is a confusion table? What does it help us understand about a model's performance?

The confusion table is important because it lets us understand the Type 1 (FP) & Type 2 (FN) errors made by the model. It also lets us calculate specific metrics like accuracy, f1 score, tpr/fpr.
   
3. What does the SSE quantify about a particular model?

SSE measures the error residuals of the y_hat in the particular model. This is particular useful for train-test to evaluate how "off" the estimate was for the particular model.

4. What are overfitting and underfitting?

Overfitting is having to many variables such that you can predict the target classes very well, but it might not generalize to the test set. Underfitting is having to little useful variables such that you might be way off to the target, but also might not generalize to the test set. This is a trade-off between model complexity, but the important thing to remember is that both are bad for the test-set and generally you need to find the "sweet spot".

(Hint: Train/test split, iterate over reasonable values of $k$ and find a value that minimizes SSE on the test split using predictions from the training set.)
5. Predict the missing ratings.

The predicted missing ratings are more dense around the larger range 80-100, so there isn't any values <40 for the predictions. Such that you won't get "low-reviews" because there aren't any closer prices/beds.

```python
import pandas as pd 

df = pd.read_csv("../data/airbnb_hw.csv")
print(df.shape)
```

```python
missing_df = df[df.isnull().any(axis=1)]
print(missing_df.shape)
```

```python
df = df.dropna(axis = 0, how = 'any')
```

```python
df["Price"] = df["Price"].str.replace(",","")
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

x = df[["Price","Beds"]]
y = df["Review Scores Rating"]

sse_scores = []
k_range = range(1,26)

for i in range(1, 26):
    model = KNeighborsRegressor(n_neighbors=i)
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    model.fit(x_train, y_train)
    y_hat = model.predict(x_test)
    squared_errors = (y_test - y_hat)**2
    sse = np.sum(squared_errors)
    sse_scores.append(sse)

plt.figure(figsize=(10, 6))
plt.plot(k_range, sse_scores, marker='o', linestyle='-', color='b')
plt.title('Elbow Plot for KNeighborsRegressor')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('SSE')
plt.grid(True)
plt.show()
```

```python
model = KNeighborsRegressor(n_neighbors=15)
model.fit(x,y)
missing_df = missing_df[~missing_df[["Price", "Beds"]].isna().any(axis=1)]
missing_df["Price"] = missing_df["Price"].str.replace(",","")
missing_df["Review Scores Rating"] = model.predict(missing_df[["Price","Beds"]])
```

```python
import seaborn as sns

fig,ax = plt.subplots(1,2, figsize=(10,5))
sns.kdeplot(df["Review Scores Rating"], ax=ax[0])
ax[0].set_title("Train")
sns.kdeplot(missing_df["Review Scores Rating"], ax=ax[1])
ax[1].set_title("Test")
```
